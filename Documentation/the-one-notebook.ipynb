```python
import numpy
```

```python
from math import pi, sin
```

```python
def mysin(x, amplitude, hz, phase_shift, vertical_shift):
    return amplitude * sin(hz*(x+phase_shift)) + vertical_shift
```

```python
mysin(2*pi, hz=10, amplitude=5, phase_shift=pi/2, vertical_shift=37)
```

```python
import matplotlib.pyplot as plt
```

```python
xs = numpy.linspace(0, 2*pi)
```

```python
xs
```

```python
plt.plot(xs, [mysin(x, hz=10, amplitude=5, phase_shift=pi/2, vertical_shift=37) for x in xs])
```
